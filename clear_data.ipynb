```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_excel('File_import_data_Nhân_viên_và_Hợp_đồng__3_P edited.xlsx', header=2)
```

```python
date_columns = ['Ngày vào làm thực tế', 'Ngày nhân viên chính thức', 'Ngày thôi việc',
                'Ngày sinh', 'Ngày sinh', 'Ngày bắt đầu', 'Ngày kết thúc']
data[date_columns]
```

```python
def convert_date_format(date_str):
    try:
        return datetime.strptime(date_str, '%d/%m/%Y').strftime('%Y-%m-%d')
    except (ValueError, TypeError):
        return date_str

# Lặp qua từng cột dữ liệu và chuyển đổi định dạng nếu cần
for column in date_columns:
    data[column] = data[column].apply(convert_date_format)
```

```python
data.to_excel('output_file.xlsx', index=False)
```
